```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data 수집
# data 갯수 : N개 64개
# 입력 데이터 갯수: Din, 4
# 출력값의 갯수 : Dout, 2

N, Din, Dout = 64, 4, 2

# 입력 데이터:x, 출력값 : y, 파라미터: w
x = np.random.randn(N, Din)
y = np.random.randn(N, Dout)
w = np.random.randn(Din, Dout)

print(x.shape)
print(y.shape)
print(w.shape)
```

## 훈련

```python
lr = 1e-6 # 학습률
loss_arr = []
for t in range(5000):
    # 예측값을 계산
    y_hat = np.matmul(x,w)

    # loss 계산
    loss = 0.5*((y_hat - y)**2).sum()
    loss_arr.append(loss)
    # 경사하강법 적용 -> w의 미분
    grad_y_hat = y_hat - y
    grad_w = np.matmul(np.transpose(x), grad_y_hat)

    # w를 갱신
    w -= lr*grad_w

plt.plot(loss_arr)
```
