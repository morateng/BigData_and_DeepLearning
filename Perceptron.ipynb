## numpy를 이용한 단순한 구현

```python
import numpy as np
import matplotlib.pyplot as plt
# N: 데이터 개수(64), Din: x의 차원(4), Dout: y의 차원(2)
# x: 64x4
# y: 64x2
# w: 4x2

N, Din, Dout = 64, 4, 2

x = np.random.randn(N, Din)
y = np.random.randn(N, Dout)
w = np.random.randn(Din, Dout)

lr = 1e-6
loss_arr = []

for t in range(50000):
    y_hat = np.matmul(x, w)  # 예측값 계산 (64x4) * (4x2) -> 64x2

    loss = (0.5*(y_hat - y)**2).sum()  # loss계산
    loss_arr.append(loss)

    grad_y_hat = (y_hat - y)  # y_hat에 대한 미분
    grad_w = np.matmul(np.transpose(x), grad_y_hat) # w에 대한 미분

    w -= lr*grad_w  # 가중치 업데이트

plt.plot(loss_arr)
```
